#### @tool

in this the structure is given by type hinting, but not strictly enforced

```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a: int, b: int)-> int:
    """multiply two numbers"""
    return a*b
```

```python
multiply.invoke({'a':3, 'b':4})
```

```python
print(multiply.name)
```

```python
print(multiply.description)
```

```python
print(multiply.args)
```

```python
# this is what our llm see

multiply.args_schema.model_json_schema()
```

#### structuredtool & pydantic

here schema is enforced very strictly using pydantic

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(..., description='The first number to multiply ')
    b: int = Field(..., description='The first number to add ')
```

```python
def multiply_fuc(a: int, b: int) -> int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_fuc,
    name='multipy',
    description='multiply two numbers',
    args_schema=MultiplyInput
)
```

```python
multiply_tool.invoke({'a': 2, 'b': 4})
```

#### Basetool class

```python
from langchain.tools import BaseTool
from typing import Type
```

```python
# arg schema using pytandic

class MultiplyInput(BaseModel):
    a: int = Field(..., description='The first number to multiply ')
    b: int = Field(..., description='The first number to add ')
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
multiply_tool.invoke({'a': 3, 'b': 3})
```

#### Toolkit

```python
@tool
def add(a: int, b: int) -> int:
    """add two numbers"""
    return a+b

@tool
def multi(a: int, b: int) -> int:
    """multi two numbers"""
    return a*b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multi]
```

```python
toolkit = MathToolkit()

tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, "=>", tool.description)
```
