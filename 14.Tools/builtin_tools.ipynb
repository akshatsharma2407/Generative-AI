# DuckDuckGo

```python
from langchain_community.tools import DuckDuckGoSearchRun
```

```python
search_tool = DuckDuckGoSearchRun()
```

```python
result = search_tool.invoke('bangladesh news on hindu today')
```

```python
print(result)
```
