```python
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
import requests
import os
from dotenv import load_dotenv
import json
```

```python
load_dotenv()
WEATHER_API = os.getenv('weather_api')
```

```python
@tool
def get_weather_data(city: str) -> str:
    """ 
    this function fetches the current weather data for given city
    """

    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city}/2025-12-28?key={WEATHER_API}"

    res = requests.get(url).content
    return json.loads(res)['days'][0]['temp']
```

```python
res = get_weather_data.invoke({'city' : 'mumbai'})
```

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os

load_dotenv()

GROQ_API_KEY = os.getenv('GROQ_API_KEY')

model = ChatGroq(
    model="llama-3.1-8b-instant"
)
```

```python
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_classic import hub
```

```python
prompt = hub.pull("hwchase17/react")
```

```python
search_tool = DuckDuckGoSearchRun()
```

```python
agent = create_react_agent(
    llm=model,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True 
)
```

```python
agent_executor.invoke({'input' : 'what is capital of argentina and what is temperature there?'})
```
